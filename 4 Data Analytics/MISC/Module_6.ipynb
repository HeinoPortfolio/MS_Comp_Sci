```python
X = range(10)
Y = [value**2 for value in X]

print(X)
print(Y)
```

```python
# Create a scatterplot of the values
# will show the trend in the data/values
plt.scatter(X, Y)
```

```python
# Adding lables to the scatterplot

work = list(df['Work'])
leisure = list(df['Leisure'])
names = list(df.index)

print(work)
print(leisure)
print(names)
```

```python
# Draw the scatterplot

plt.scatter(work, leisure)
plt.xlabel("Work")
plt.ylabel("Leisure")
plt.title("Work vs. Leisure")

# Add labels
for name, work_hours, leisure_hours in zip(names, work, leisure):
    plt.annotate(name, xy=(work_hours, leisure_hours), xytext=(10,10), textcoords='offset points'    )
```

## Area Chart

```python
# Creating an Area Chart

x=range(1,6)
y=[1,5,7,8,3]

plt.fill_between(x, y, color='purple')
plt.show()
```

# Stacked Area Chart

```python
x=range(1,6)
y1 = [1, 2, 6, 8, 9]
y2 = [2, 2, 7, 10, 12]
y3 = [2, 8, 5, 10, 6]

plt.stackplot(x, y1, y2,y3, labels=['A','B','C'])
plt.legend(loc='upper left')
plt.show()
```
