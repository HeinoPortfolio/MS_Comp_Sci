## Performing Visualizations on Time Series Data


```python
# Create 5 datetime objects
dates = [datetime(2018, 10, 1),datetime(2018, 10, 2),datetime(2018, 10, 5),datetime(2018, 10, 8),datetime(2018, 10, 10)]

dates
```

```python
# List of values
levels = [1.5, 1.8, 2.3, 1.9, 1.3]
```

```python
from pandas import Series
```

```python
ts = Series(levels, index = dates)

ts
```

```python
# Creating a line chart with the Time Series data
ts.plot.line(marker='o', linestyle='-')
```

```python
# Draw a bar chart treating each data as a --- label.
ts.plot.bar()
```

### Using date_range()

```python
import pandas as pd
```

```python
pd.date_range('8/1/2018', '8/31/2018')
```

```python
Aug_2018 = pd.date_range('8/1/2018', '8/31/2018')
    
Aug_2018
```

```python
sea_levels = [2, 25, 48, 28, 5, 24, 23, 28, 5, 53, 25, 89, 1, 25, 29, 21, 13, 29, 13, 29, 29, 27, 17
              , 28, 48, 55, 58, 75, 68, 63, 39,]
```

```python
from pandas import DataFrame

# Create a pandas dataframe with two columns (dates and sea levels)
df = DataFrame({'Dates': Aug_2018, 'Level': sea_levels})

df.reset_index(drop=True, inplace=True)
```

```python
df.head(10)
```

```python
df.plot.line(x='Dates', y='Level')
```

### Moving Window Creation

```python
df['Moving Avg'] = df['Level'].rolling(window=3).mean()

df
```

```python
# Create a comparison graph of time seris data vs moving averages (mean).
df.plot.line(x='Dates', y=['Level','Moving Avg'])
```

```python
df['Moving High'] = df['Level'].rolling(window=3).max()

df.head(10)
```

```python
# Create a comparison graph of time seris data vs moving highes.
df.plot.line(x='Dates', y=['Level','Moving High'])
```

```python
df['Moving Low'] = df['Level'].rolling(window=3).min()

df.head(10)
```

```python
# Create a comparison graph of time seris data vs moving highs.
df.plot.line(x='Dates', y=['Level','Moving Low'])
```

```python
df['Moving Std'] = df['Level'].rolling(window=3).std()

df.head(10)
```

```python
# Create a comparison graph of time seris data vs standard deviation.
df.plot.line(x='Dates', y=['Level','Moving Std'])
```

```python
df.plot.line(x='Dates', y=['Level','Moving Avg', 'Moving Std'])
```
