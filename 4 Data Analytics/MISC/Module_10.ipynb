```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
```

```python
# Using the BSU Web page
response = urlopen("file:bsu_data_science.html")

html = response.read()
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
#print(soup.prettify())  # uncomment to view full htlm code page
```

### Extract content from the 'p' tags of the web page

```python
contents = []

# find all the paragraph tags and add paragraphs to a list
for p in soup.find_all('p'):
    contents.append(p)

len(contents)
```

```python
# Print one paragraph's content
print(contents[1])
```

```python
# Display the content associated with each paragraph
for content in contents:
    print(content.text)
    print('\n')
```

### Using a JSON for Tweets

```python
import json
```

```python
with open('one_tweet.json', 'r') as json_file:
    dict_2 = json.load(json_file)
    
# how many key-value pairs
print(len(dict_2))
```

```python
# Show the keys
print(dict_2.keys())
```

```python
# show the values
print(dict_2.values())
```

```python
# Use a for loop to traverse the dictionary
for k in dict_2.keys():
    print(k,'\n', dict_2[k], '\n ==========')
```
