```python
# Load the positive words from the text file
pos_dict = []

f = open("positive_words_list.txt")

f.seek(0)

for line in f:
    
    word_pos = line.strip()
    pos_dict.append(word_pos)

# Close the file
f.close()
```

```python
print(len(pos_dict))

# Print some of the contents of the positive wards list
print(pos_dict[:20])

print(pos_dict[100:150])
```

```python
# Load the negative words from the text file
neg_dict = []

f = open("negative_words_list.txt")

f.seek(0)

for line in f:
    
    word_neg = line.strip()
    neg_dict.append(word_neg)

# Close the file
f.close()
```

```python
print(len(neg_dict))

# Print some of the contents of the positive wards list
print(neg_dict[:20])

print("\n\n",neg_dict[1000:1050])
```

```python
# Remove the punctuation from the lines of comments

for index in range(len(comments)):
    
    comments[index] = comments[index].translate(str.maketrans('','',string.punctuation))
    
# Show the comments after removal
comments
```

```python
# Count negative and positive words

positive_score = 0
negative_score = 0

matched_pos_words = []

# Look at the first comment in the list and split into tokens.
comment_words = comments[0].split()

for word in comment_words:
    
    word = word.lower()
    
    if word in pos_dict:
        positive_score = positive_score + 1
        matched_pos_words.append(word)
   
    if word in neg_dict:
        negative_score = negative_score + 1

print(positive_score)
print(negative_score)
```
