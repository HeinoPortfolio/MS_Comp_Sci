```python
import warnings

warnings.filterwarnings("ignore")
```

```python
l= ['B','A','B','C','C','A','A', 'B','A']

print(l)
```

```python
s = set(l)

print(s)
```

```python
# Create a list of keys
keys = list(s)

print(keys)
```

```python
# Create a dictionary with keys.
dict = {}
for key in keys:
    dict[key] = 0
    
print(dict)
```

```python
# Print the keys and values
print(dict)
```

```python
#print the keys
print(dict.keys())
```

```python
# Traverse the list to find the occurneces of each key

for key in dict.keys():
    for v in l:
        if key == v:
            dict[key] = dict[key] + 1
            
# Print the results
print(dict)
```

```python
# To see the values
print(dict.values())
```

## Create a bar chart

```python
import matplotlib.pyplot as plt
```

```python
# Draw a bar chart
X = ['A','B','C','D','F']
Y = [6, 12, 8, 4, 2]

plt.bar(X, Y)
plt.xlabel('Grades')
plt.ylabel('Counts')
plt.title("Grades Distribution")
```

```python
# Horizontal bar chart

plt.barh(X, Y)
plt.xlabel('Counts')
plt.ylabel('Grades')
plt.title("Grades Distribution")
```

```python
## Create a pie chart

plt.pie(Y, labels = X)
```

```python
# Display percentage of each category

color_set = ['#ff0000', '#00ff00', '#0000ff', '#ffffAA', '#BC005A']
explode = (0, 0.1, 0, 0 ,0)

plt.pie(Y, colors=color_set, labels=X, autopct='%1.1f%%', shadow=True, startangle=90)
```

```python
# Emphasize a particular segment

color_set = ['#ff0000', '#00ff00', '#0000ff', '#ffffAA', '#BC005A']
explode = (0, 0.1, 0, 0 ,0)

plt.pie(Y, colors=color_set, labels=X, autopct='%1.1f%%', shadow=True, startangle=90, explode=explode)
```

## Donut chart

```python
plt.pie(Y, labels=X)
my_circle=plt.Circle((0,0), 0.6, color='white')

p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
tips = pd.read_csv("tips.csv")

len(tips)
```

```python
tips.head(10)
```

```python
# Group the data by day then by sex column --- TWO-LEVEL GROUPING
tip_summary = tips.groupby(['day','sex'])['tip'].sum().unstack()

print(tip_summary)
```

## Stacked Bar Chart

```python
tip_summary.plot(kind='bar', stacked=True)

plt.title("Tip by Day and Gender")
```

```python
# Horizontal

tip_summary.plot(kind='barh', stacked=True)

plt.title("Tip by Day and Gender")
```

```python
# Regular Bar chart

tip_summary.plot(kind='bar')

plt.title("Tip by Day and Gender")
```
