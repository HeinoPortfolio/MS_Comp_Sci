### Dictionary Demo
- **.keys()** - will return all the keys in the dictionary.
- When adding a new value to the dictionary make sure the key is **unique**.  Cannot have the same key values.
- To update a dictionary supply the key and the new value for the key.

```python
dict = {'John':14, 'Tom': 18, 'Susan':17, 'Helen':19}
```

```python
len(dict)              # will return the length of the dictionary
```

```python
dict.keys()             # will return the keys in the dictionary
```

```python
# processing a dictionary using a for loop
for key in dict.keys():
    print(key)                         # prints the key
    print(dict[key])                   # prints the value
    print('-----')
```

```python
# to add a new value to the dictionary
dict['Andy'] = 13
dict
```

```python
# to update a key and its value
dict['John'] = 24
dict
```

### Creation of a DataFrame object

**Notes**:
- Each in a dictionary will be the name of the column in the dataframe.
- Adding one more pair of [] will return a DataFrame.
- **.index** - to the range of the indexes of the dataframe
- **.columns** - to view the columns of the dataframe
- **.head()** - to view a specific number of the top rows, default is 5
- **.tail()** - to view a specific number of the bottom rows, default is 5
- **.iloc[]** - to access data via the index in the dataframe 
- **.loc[]** - to access data via the label index in the dataframe


```python
# Creates a dictionary
data = {'state': ['Ohio','Ohio', 'Ohio', 'Nevada', 'Nevada'], 'year':[2000, 2001, 2002, 2001, 2002], 'pop':[1.5, 1.7, 3.6
                                                                                                             ,2.4, 2.9]}
```

```python
type(data)
```

```python
data
```

```python
# Create a dataframe out of the dictionary (data)
df1 = DataFrame(data)

df1
```

```python
# to view the indexes of the DataFrame
df1.index
```

```python
# to view the columns of the dataframe
df1.columns
```

```python
# to view a specific number of top rows
df1.head(2)
```

```python
# to view a specific number of bottom rows
df1.tail(3)
```

```python
# to assign own lables to the rows 
df1.index = ['one','two','three','four','five']
df1
```

```python
# to access data from the dataframe
df1.iloc[0]                              # will get the first column of the dataframe
```

```python
# Checking the type
type(df1.iloc[0])
```

```python
# adding one more pair of []
df1.iloc[[0]]
```

```python
type(df1.iloc[[0]])
```

```python
# to access data from the dataframe using the label index
df1.loc['one'] 
```

```python
# adding one more pair of []
df1.loc[['one']]
```

```python
type(df1.loc[['one']])
```

```python
# to retrieve a specific column
df1['state']
```

```python
type(df1['state'])
```

```python
df1
```

```python
df1[['state']]         # returns a dataframe not a Series
```

```python
# to access a particular cell
df1.iloc[0, 2]                          #  gives the first row and the third column
```

```python
# using label indexe
df1.loc['one', 'pop']                  #  gives the first row and the third column
```

```python
# For a range or a segment (a slice of the dataframe)
df1.iloc[0:2, 0:2]
```

```python
# to display values not in a fixed range use the labels
df1.loc[['one', 'five'], ['state','pop']]
```

```python
df1.iloc[[0,4], [0, 2]]
```
