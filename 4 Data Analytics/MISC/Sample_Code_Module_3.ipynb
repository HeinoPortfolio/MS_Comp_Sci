### Dictionary Demo
- **.keys()** - will return all the keys in the dictionary.
- When adding a new value to the dictionary make sure the key is **unique**.  Cannot have the same key values.
- To update a dictionary supply the key and the new value for the key.

```python
dict = {'John':14, 'Tom': 18, 'Susan':17, 'Helen':19}
```

```python
len(dict)              # will return the length of the dictionary
```

```python
dict.keys()             # will return the keys in the dictionary
```

```python
# processing a dictionary using a for loop
for key in dict.keys():
    print(key)                         # prints the key
    print(dict[key])                   # prints the value
    print('-----')
```

```python
# to add a new value to the dictionary
dict['Andy'] = 13
dict
```

```python
# to update a key and its value
dict['John'] = 24
dict
```

### Sets Demo
- **set()** - creates a set from the given values

```python
set1 = set([13, 5, 6, 8])
set2 = set([2, 4, 6, 8])
```

```python
print(set1)
print(set2)
```

### Union Demo
- **union()** - joins two sets together.
- If there are duplicates only one element will be kept in the new union.

```python
set1.union(set2)
```

```python
# intersection example
set1.intersection(set2)
```

```python
# difference example
set1.difference(set2)
```

```python
set2.difference(set1)
```

```python
from pandas import Series, DataFrame
```

```python
# creating a Series object from a list of integers.
obj1 = Series([4, 7, -5, 3])
```

```python
type(obj1)
```

```python
obj1
```

```python
obj1[2]         # to access an element by its index
```

```python
obj1.index     # will show the range of the indexes for the series
```

```python
obj1.values   # will display the values of the series
```

```python
# assigning index names to the series
obj1.index = ['d','b','a','c']
obj1
```

```python
# to access the content of a series using the label
obj1['a']
```

```python
# to access the content of a series using the index
obj1[2]
```

```python
# assigning index names at the same time as creating the Series
obj2 = Series([4, 7, -5, 3], index=['d','p','a','h'])
obj2
```
