## Perform Machine Learning

```python
# Use half of the data to train
xtrain = data_1[0:21000, 1:]

# Save the data labels
train_label = data_1[0:21000, 0]
```

```python
train_label[:10]
```

```python
# Create a Decision Tree Classifier
clf = tree.DecisionTreeClassifier()
```

```python
#Train/ learn the data to create a decision tree for future predictions
clf.fit(xtrain, train_label)
```
