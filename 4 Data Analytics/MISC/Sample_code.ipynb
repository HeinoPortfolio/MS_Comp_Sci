```python
l.sort()
l
```

## Dictionary Demo

## Notes
- Keys must be unique /unique!
- **.keys()** will return a list of the keys in the dictionary
- **.values** will return a list of the values in the dictionary

```python
dict = {'John':14, 'Tom':18, 'Susan':17, 'Helen': 19}
print(len(dict))
```

```python
dict.keys()      # will return all the keys in the dictionary
```

```python
dict.values()   # will return all the values in the dictionary
```

```python
# use the for loop to iterate through a dictionary
for k in dict.keys():
    print(k)
    print(dict[k])
    print('----')
```
