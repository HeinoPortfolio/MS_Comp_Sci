```python
import pandas as pd

# Pre-question stuff

df = pd.read_csv('video_rental.csv')

print(df)
```

# Question 1

How many rows and column does this dataset have?

```python
print("Question 1:\n \nNumber of rows and columns: ", df.shape)
```

# Question 2

How many male and female customers in this dataset?

```python
by_gender = df.groupby("Gender")['Gender'].count()

print("\nQuestion 2:\n \nNumber by male and female customers: \n", by_gender)
```

# Question 3 

What is the average income of the female customers.

```python
avg_inc_female = df.groupby("Gender")['Income'].mean()
print("\nQuestion 3:\n \nAverage income of female customers: \n", avg_inc_female)
```

```python
avg_inc_female2 = df[df['Gender'] == 'F']
avg_inc_fem = avg_inc_female2['Income'].mean()

print("\nQuestion 3 (version 2):\n \nAverage income of female customers: ", avg_inc_fem)
```

# Question 4

For male customers who love Drama genre, what is the oldest age?

```python
drama_male = df[(df['Gender'] == "M") & (df['Genre'] == 'Drama')]['Age'].max()

print("\nQuestion 4:\n \nThe oldest age of a male that likes drama: ", drama_male)
```

# Question 5

What is the youngest age of male customers?

```python
youngest_male = df[df['Gender'] == 'M']['Age'].min()
print("\nQuestion 5: \nThe youngest age of male customers: ", youngest_male)
```

# Question 6

What is the average age among female customers who love Drama movies?

```python
drama_female = df[(df['Gender'] == "F") & (df['Genre'] == 'Drama')]['Age'].mean()
print("\nQuestion 6: \nThe average age among female customers who love drama: ", drama_female)
```

# Question 7

Use the **Genre** column to draw a bar chart.

```python

bar_genre = df['Genre'].value_counts().plot(kind='bar',title='Frequency of Each Genre', ylabel='Frequency').get_figure()
bar_genre.savefig('bar_genre.pdf')
```

# Question 9 

Use the Genre column to draw a pie chart.

```python
count_1 =  df.groupby(['Genre'])['Genre'].count()
print(count_1)

count_img = df.groupby(['Genre'])['Genre'].count().plot(kind='pie', autopct='%1.0f%%', title='Percentage of Each Genre'
                                                        ,ylabel="").get_figure()

count_img.savefig('pie_genre.pdf')
```

# Question 10

Use the Genre, Gender, and Age columns to draw a bar chart that shows the average ages of Male and Female customers separately on each movie genre.

```python
genre_summary = df.groupby(['Genre','Gender'])['Age'].mean().unstack()

avg_img = genre_summary.plot(kind='bar', title='Average Age of Customer by Genre and Gender').get_figure()

avg_img.savefig('avg_genre.pdf')
```
