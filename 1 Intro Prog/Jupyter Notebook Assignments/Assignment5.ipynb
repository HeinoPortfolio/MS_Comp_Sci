#### Second Task: Calculating the average 

Now, use this list to calculate the average of all numbers entered.  In this case, you are **required to use a for loop** when doing the calculation (this is one of the learning objectives for this assignment!).

If you look at the code below, a function has been outlined for you, with two areas for you to add code to calculate the average.  First, you will add up all the numbers in the list.  Then you will calculate the average by dividing by 5 (how many numbers are in the list).

**Hint:** It is very common for beginning programmers to want to use print() instead of return.  Here, we _don't_ want this function to actually print the result -- we just want it to **return** the right answer, so that another section of code could print it, if we wanted to.

Write your code in the following cell, execute the cell, and then scroll down to the unit tests that the autograder will use -- and execute them to see if your solution to this function is correct.

```python
# Use a for loop (with range()) to add up all of the grades in the list. 
# We have created a function that takes the list of numbers as parameter 
# and contains a variable named average already. Add your code in the mentioned spaces

def calc_average(grade_list) :

    average = 0
    
    # REMEMBER to remove the `raise NotImplementedError()` statement when you
    # have added your code.  If you do not remove this statement, your code
    # will fail the autograding tests.
    # (This line is included here to ensure the functioning of the autograder.)
    total = 0
    
    for index in range(0, 5):
        total = total + grade_list[index]
    
    ## After the for loop finishes, divide the total by 5, in order to calculate the average.
    
    average = total / 5 
    
    return average
```
