```python
same_list = ['Ford', 'BMW', 'Volvo']  # list 
same_tuple = ('Ford', 'BMW', 'Vovlo') # tuple

print('Length of list: ', len(same_list), "Length of tuple: ", len(same_tuple))
print('Slicing a list', same_list[:-1])
print('Slicing a tuple', same_tuple[:-1])

# Sorting a list and a tuple
print(sorted(same_list))
print(sorted(same_tuple))
```

```python
search_list = ['Ford', 'Chevy', 'Hyundai', 'Vovlo']

for word in search_list:
    # see if word is in the list
    if word in same_list:
        print(word, " is in the the list.")
    if word not in same_list:
        print(word, "is NOT in the list.")

```

```python
import random

random.seed(1)
number_list = []
for number in range(0,100):
    number_list.append(random.randint(0, 6))
    
print(number_list)
for num in range(0,7):
    print('Number: ', num, ' appeared:', number_list.count(num))
    
```
