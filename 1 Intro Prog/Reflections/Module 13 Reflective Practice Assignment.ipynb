```python
import matplotlib.pyplot as plt
import pandas as pd
import nltk
#import string

from operator import itemgetter
from nltk.corpus import stopwords
from pathlib import Path
from textblob import TextBlob
from wordcloud import WordCloud

#nltk.download('stopwords')
#nltk.download('punkt')
```

```python
raw_text = Path('JulesVerne.txt').read_text()
#print(raw_text)

# create the text blob
verne_blob = TextBlob(raw_text)

# load the stopwords
stop_words = stopwords.words('english')

# Create an variable withteh word counts 
word_counts = verne_blob.word_counts.items()
#print(word_counts)

# remove the stopwords.
no_stopwords = [each_word for each_word in word_counts if each_word[0] not in stop_words]
#print(no_stopwords)

# sort the result with the stopwords
sorted_words = sorted(no_stopwords, key=itemgetter(1), reverse=True)
#print(sorted_words)

without_punct = []
punctuation = ['“', '”', "’"]

# Remove the the punctuation that should not be included.
for word in sorted_words:
    if word[0] not in punctuation:
        #print(word)
        # add to the list
        without_punct.append(word)

#print(without_punct)

# top 20 words
top20 = without_punct[0: 20]
print(top20)
#print(len(top20))




```

```python
# create the word cloud

# using a WordCloud visualize word cloud
wordcloud = WordCloud(colormap='prism', max_words=20)
wordcloud = wordcloud.generate(raw_text, )


# show the word cloud
# plot the WordCloud image                       
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()


```
