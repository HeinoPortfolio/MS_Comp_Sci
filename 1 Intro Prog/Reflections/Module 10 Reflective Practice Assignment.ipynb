```python
# use  different methods on a string. ( .strip(), .lower(), .upper(), .capitalize(), .title(), .count(),  .replace())

user_string = input('Please enter a string (type "quit" no quotes to exit)')

while user_string != 'quit':
    print("***", user_string, "***")
   
    # remove the whitespace 
    print(user_string.strip())
    
    print("Lower case: ", user_string.lower())
    print("Upper case: ", user_string.upper())
    
    print("Capitalized: ", user_string.capitalize())
    print("Using title: ", user_string.title())
    
    print("Count: ", user_string.count('a'))
    
    print("With the letter a replace: ", user_string.replace('a', 'abc'))
    
    # get new string from the user
    user_string = input('Please enter a string (type "quite" no quotes to exit)')
    
print('Program finished.')
```
