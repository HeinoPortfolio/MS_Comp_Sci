```python
num_to_guess = 23 
while True:
    num_guessed = int(input('Please enter a guess: '))
    if num_guessed > num_to_guess:
        print('Guess is too high!')
    elif num_guessed < num_to_guess:
        print('Guess is too low!')
    else:
        print('You guessed the number. Way to go!')
        break
```
