```python
# part one 
import re

from textblob import TextBlob

import nltk
nltk.download('brown')
nltk.download('punkt')

# prompt the use for a long string and do the sentiment analysis and extract  
user_str = input("Please enter a sentence:")
#print(user_str)

# extract phone number from the string
pattern = pattern = '\d{3}-\d{3}-\d{4}'
phone_number = re.findall(pattern, user_str)
print("Phone_number: ", phone_number)

# sentiment analysis
string_blob = TextBlob(user_str)
print(f"Sentiment in the string is: {string_blob.sentiment.polarity:.1f}")

# show an emoji
grinning_face = 0x1F600 # regular grinning face
neutral_face = 0x1F610

if string_blob.sentiment.polarity == 1:
    print(f"Great {grinning_face:c} ")
else:
     print(f"Great {neutral_face:c} ")
        
# word count on the string
print(string_blob.noun_phrases)

# remove all whitespace from a string
new_str = ''
new_str = "".join(user_str.split())
print(new_str)

# change string into a list of characters
char_list =  list(new_str)
print(char_list)


lst  = map(lambda ch:ch.lower(), char_list)
print(lst)

sorted_list = sorted(lst)



print(sorted_list)
```
