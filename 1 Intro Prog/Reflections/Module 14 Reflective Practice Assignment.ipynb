```python
import requests
import json

article = 'Michael_Jordan'
url = url = 'https://en.wikipedia.org/w/api.php?action=parse&page='+ article +'&format=json'

print(url)
```

```python
response = requests.get(url)
print(response)
```

```python
#print(response.json())
michael_jordan_str = json.dumps(response.json(), indent=4)
#print(michael_jordan_str)

michael_jordan = response.json()
#print(michael_jordan)
print('Title: ', michael_jordan['parse']['title'])
```

```python
filename = 'dummyjsonexample.txt' 
#json_file1 = json.load(filename)

with open(filename, 'r') as file:
    file = file.read()
    
#print(file)
#print(type(file))

# convert to a JSON object
file1_json = json.loads(file)

#print(file1_json)
#print(type(file1_json))

print (file1_json.keys())

print("\nID: ", file1_json['id'])
print("\nTitle: ", file1_json['title'])
print("\nDescription: ", file1_json['description'])
```
