```python
#There are 20 points in this demo like the following
points = [[52, 28],[25, 95], [36, 64],[20, 94],[52, 27],[20, 96],[39, 65],[34, 90],[75, 15],
[74, 21],
[90, 43],
[60, 12],
[67, 20],
[30, 92],
[61, 31],
[8, 74],
[51, 32],
[64, 41], [15, 80], [9, 80]]
```

### Functions used for visualization  

- extracts the x and y values from the pair

```python
#used for visualization

def get_X(P):
    r = []
    for x in P:
        r.append(x[0])
    return r
```

```python
def get_Y(P):
    r = []
    for x in P:
        r.append(x[1])
    return r
```

## Plot a graph of the pairs of points

```python
import matplotlib.pyplot as plot

%matplotlib inline

plot.ylim([0,105])
plot.xlim([0,105])

X = get_X(points)
Y = get_Y(points)

plot.scatter(X, Y)
plot.show()
```

### Import KMeans from scikit learn.

```python
from sklearn.cluster import KMeans
```

```python
import numpy as np
```

```python
# Convert list of points to numpy array

X = np.array(points)
```

```python
### Find two clusters of X.
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
```

### Show the cluster labels; starting from 0.

```python
kmeans.labels_
```

### Convert the array of labels to a list.

```python
cluster_labels = list(kmeans.labels_)

cluster_labels
```

### Create two clusters.

```python
cluster_1_points_index = []

cluster_2_points_index = []

for i in range(len(cluster_labels)):
    
    if cluster_labels[i] == 0:
        cluster_1_points_index.append(i)
    elif cluster_labels[i] == 1:
        cluster_2_points_index.append(i)
    else:
        print("Something went wrong.")
        break
```

```python
# verification

cluster_1_points_index
```

```python
### Group cluster 1 points.
```

```python
cluster_1_points = []

for x in cluster_1_points_index:
    cluster_1_points.append(points[x])

cluster_1_points
```

```python
cluster_2_points_index
```

```python
### Group cluster 2 points
```

```python
cluster_2_points = []

for x in cluster_2_points_index:
    cluster_2_points.append(points[x])

cluster_2_points
```

```python
# Visualize two clusters. 
```

```python
import matplotlib.pyplot as plot

X = get_X(cluster_1_points)
Y = get_Y(cluster_1_points)

plot.ylim([0,105])
plot.xlim([0,105])

plot.scatter(X, Y, color="green")

X = get_X(cluster_2_points)
Y = get_Y(cluster_2_points)

plot.ylim([0,105])
plot.xlim([0,105])

plot.scatter(X, Y, color="red")

plot.show()
```
