## Simple data structures

```python
# list is a data structure to store values
l = [6, 7, 2, 1, 4]

print(l)
print(len(l))
```

```python
# show the first element of l
print(l[0])
```

```python
# show the last element of l
print(l[len(l) - 1])
```

## For loop

For traversing a list

```python
for i in range(len(l)):
    print(l[i])
```

```python
for x in l:
    print(x)
```

```python
# sort the list
l.sort()

print(l)
```

```python
# descending order
l.sort(reverse=True)

print(l)
```

```python
# sort strings
# will be sorted by the lexical order

s_1 = ['banana', 'orange', 'apple']

print(s_1)

s_1.sort()
print(s_1)
```

```python
# lists within lists: 2-dimensions

l_two = [[23, 3, 1], [6, 45, 2, 5], [21, 456]]

print(l_two)
```

```python
# iterate through the list use two for loops

for x in l_two:
    for y in x:
        print(y)
    print("----------")
```

# Functions

```python
def my_sum(l):
    sum = 0
    for x in l:
        sum = sum + x
    return sum
```

```python
# calling the function

print(l)

my_sum(l)
```

```python
dict = {'John': 14,'Tom': 18, 'Susan': 17, 'Helen': 19}
print(len(dict))
```

```python
# will read a file line-by-line into a list

fin = open("D:\\Jupyter_Notebooks\\CS654ML_DM\\Data\\sample_input.txt", 'r')
data_raw = []
for l in fin:
    data_raw.append(l)

fin.close()
```

```python
data_raw
```

```python
# using the strip function
data_int = []

for i in range(len((data_raw))):
    data_int.append(int(data_raw[i].strip()))

data_int
```

# Write data to a file

```python

fout = open("D:\\Jupyter_Notebooks\\CS654ML_DM\\Data\\sample_output.txt", 'w')

for v in data_int:
    fout.write(str(v) + '\n')
    
fin.close()
```
