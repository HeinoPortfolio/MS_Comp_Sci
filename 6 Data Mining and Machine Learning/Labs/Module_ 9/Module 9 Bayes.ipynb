```python
import pandas as pd
import numpy as np
```

### Load the training data; same as presented in one of our class slides.

- Age attribute: <=30 as  1, 31..40 as 2, > 40 as 3
- Income attribute: low as 1, medium as 2, high as 3
- Student attribute: yes as 1, no as 2
- Credit attribute: fair as 1, excellent as 2
- Buy computer attribute: Yes as 1, No as 0


```python
df = pd.read_csv("D:\\Jupyter_Notebooks\\CS654ML_DM\\Data\\cs654_computers.csv")

df
```

### This is the new customer that we are going to predict a potential computer purchase

- age <= 30
- income = medium
- student = yes 
- credit_rating = fair

thus, 1, 2, 1, 1

```python
new_customer = [1, 2, 1, 1]
```

```python
df.columns
```

### Class labels for training data

```python
y = df['buy'].to_numpy()
```

### Use the first 4 attributes as training data

```python
X = df[['age', 'income', 'student', 'credit']].to_numpy()
```

### Import the Native Bayes classifier for categorical values.

```python
from sklearn.naive_bayes import CategoricalNB
```

```python
clf = CategoricalNB(force_alpha=True)
```

### Use X as training data with y class labels

```python
clf.fit(X, y)
```

### Predict if the new customer is going to buy a new computer: 

- 1:Yes, 0:No.

```python
clf.predict([np.array(new_customer)])
```
