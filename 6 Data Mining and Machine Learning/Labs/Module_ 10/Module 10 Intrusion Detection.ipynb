```python
import pandas as pd
```

```python
df = pd.read_csv("D:\\Jupyter_Notebooks\\CS654ML_DM\\Data\\kddcup.data_10_percent.txt", header=None)
```

### Use the first 1000 rows and attributes 4 to 41 in class normal.

```python
df.loc[df[41] == 'normal.'].iloc[:1000, 4:41]
```

### Put these 1000 rows in df_normal_1000. This represents normal network data.

```python
df_normal_1000 = df.loc[df[41] == 'normal.'].iloc[:1000, 4:41]
```

### Choose the first 1000 rows of neptune class.

```python
df.loc[df[41] == 'neptune.'].iloc[:1000, 4:41]
```

### Put these 1000 rows in df_neptune_1000. This represents intrusion data.

```python
df_neptune_1000 = df.loc[df[41] == 'neptune.'].iloc[:1000, 4:41]
```

### This will be used in constructiong the two-class classifier later.  

```python
df = pd.concat([df_neptune_1000, df_normal_1000])
```

### Create the class labels. 

- 0 is for normal
- 1 is for intrusion.

```python
label_1 = []

for i in range(1000):
    label_1.append(1)
    
label_0 = []

for i in range(1000):
    label_0.append(0)

labels = label_1 + label_0
```

```python
labels[:10]
```

```python
labels[1000:1010]
```

```python
len(df)
```

## Convert data into numpy array.

```python
X = df.to_numpy()
```

```python
X.shape
```

### Convert labels into numpy array.

```python
import numpy as np

y = np.array(labels)
```

```python
y.shape
```

### Split the data into half training and half testing.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, random_state = 42)
```

### Do verification.

```python
print(len(X_train))

print(len(X_test))

print(len(y_train))

print(len(y_test))
```

### Load the knn classifier in scikit-learn

```python
from sklearn.neighbors import KNeighborsClassifier
```

### Use 1NN

```python
neigh = KNeighborsClassifier(n_neighbors=1)
```

### Train the classifier.

```python
neigh.fit(X_train, y_train)
```

### Predict the test data.

```python
predictions = neigh.predict(X_test)
```

```python
len(predictions)
```

### Check the performance.

```python
matched = 0

for i in range(1000):
    if predictions[i] == y_test[i]:
        matched = matched + 1

print(matched)
```

### Use 31 NN to train the classifier first.

- Do the prediction next.

```python
neigh = KNeighborsClassifier(n_neighbors=31)

neigh.fit(X_train, y_train)

predictions = neigh.predict(X_test)
```

### Check the performance.

```python
matched = 0

for i in range(1000):
    if predictions[i] == y_test[i]:
        matched = matched + 1

print(matched)
```
