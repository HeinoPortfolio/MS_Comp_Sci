# Basic Data Cleaning

```python
# separate each course title as a single string

courses_entries = []

for list_elem in courses_raw:
    
    raw = list_elem.strip().split(",")
    
    course_list = []
    
    for elem in raw:
        course_list.append(elem.strip())
        
    courses_entries.append(course_list)
```

```python
courses_entries
```

### Find the unique courses taken by the students

```python
unique_courses = []

for course_list in courses_entries:
    for course_name in course_list:
        if course_name not in unique_courses:
            unique_courses.append(course_name)
```

```python
unique_courses
```

```python
# order the courses in the lexical manner

unique_courses.sort()

unique_courses
```

```python
# Create an array of 1/0; 1 represents course existence, 0 means no

final_table = []

for i in range (len(courses_entries)):

    result = []
    for name in unique_courses:
        if name in courses_entries[i]:
            result.append(1)
        else:
            result.append(0)
            
    final_table.append(result)
```
