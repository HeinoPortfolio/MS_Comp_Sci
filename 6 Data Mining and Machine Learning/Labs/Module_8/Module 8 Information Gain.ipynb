```python
import math
```

```python
def info(x, y):
    
    x = float(x)
    y = float(y)
    
    if x == 0 or y == 0:
        return 0
    else:
        return (-1)* (x/(x+y))* math.log(x/(x+y), 2 ) + (-1)* (y/(x+y))* math.log(y/(x+y), 2 )
```
