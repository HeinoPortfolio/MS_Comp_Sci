## Distance computation for numerical values

```python
# assign ten integersto list l.

l = [29, 27, 15, 27, 24, 17, 35, 18, 11, 24]
```

```python
import copy
```

```python
# create a deep copy of l, and assign it to k.
k = copy.deepcopy(l)
```

```python
# sort k list.
k.sort()
```

```python
# verify k.
print(k)
```

```python
# assign the min value of k as min_v, and the max value of k as max_v.
min_v = k[0]
max_v = k[-1]

# verify the min and max values.
print(min_v)
print(max_v)
```

```python
# l is not changed. 
print(l)
```

```python
l_nor = []

for x in l:
    print(round((x-min_v)/(max_v - min_v), 2))
    l_nor.append(round((x-min_v)/(max_v - min_v), 2))
```

```python
# verify l_nor.
print(l_nor)
```
