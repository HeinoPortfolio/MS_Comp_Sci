# Distance computation for numerical values

### Assign ten integers to list l.

```python
l = [29, 27, 15, 27, 24, 17, 35, 18, 11, 24]
```

```python
import copy
```

### create a deep copy of l, and assign it to k.

```python
k = copy.deepcopy(l)
```

### Sort k list.

```python
k.sort()
```

### Verify k.

```python
print(k)
```

### l is not changed. 

```python
print(l)
```

### Assign the min value of k as min_v, and the max value of k as max_v

```python
min_v = k[0]
max_v = k[-1]
```

### Verify the min and max values.

```python
print(min_v)

print(max_v)
```

## Normalize the values

- in l into range [0..1]

```python
l_nor = []

for x in l:
    
    print(round((x-min_v)/(max_v - min_v), 2))
    
    l_nor.append(round((x-min_v)/(max_v - min_v), 2))
```

### Verify l_nor.

```python
print(l_nor)
```

### Contrast to values in l.

```python
print(l)
```

### Create empty distance_matrix and distace_row.

```python
distance_matrix = []

distance_row = []
```

## Compute the distance_matrix.

```python
import math
```

```python
distance_matrix = []

for i in range(len(l_nor)):
    
    distance_row = []
    
    for j in range(len(l_nor)):
        
        print(round(math.fabs(l_nor[i] - l_nor[j]), 2), end="\t")
        
        distance_row.append(round(math.fabs(l_nor[i] - l_nor[j]), 2))
        
    print("\n")
    
    distance_matrix.append(distance_row)
```

## Since the matrix is symmetric, thus only need to compute the upper half of it.

```python
for i in range(len(l_nor)):
    for j in range(len(l_nor)):
        #print(str(i) + ", " + str(j))
        if (j>=i):
            print(round(math.fabs(l_nor[i] - l_nor[j]), 2), end="\t")
        else:
            print("", end = "\t")
    
    print()
```

```python
### verify the distance matrix.
```

```python
# verify the distance matrix.

print(distance_matrix)
```

## Similarity matrix = 1 - distance matrix

```python
similarity_matrix = []

for i in range(len(l_nor)):
    similarity_row = []
    for j in range(len(l_nor)):
        similarity_row.append(round(1- distance_matrix[i][j], 2))
    #print("\n")
    similarity_matrix.append(similarity_row)
```

```python
# verify the similary matrix

print(similarity_matrix)
```
