## Compute the distance of each column and then add four together with weights sum up to 1.

```python
platform = ten_games['Platform'].to_list()

platform
```

## Compute the distance matrix on platform column.

```python
import math

distance_matrix_platform = []

for i in range(len(platform)):
    
    distance_row = []
    
    for j in range(len(platform)):
        #print(str(i) + ", " + str(j))
        if platform[i] == platform[j]:
            distance = 0
        else:
            distance = 1
        distance_row.append(round(distance))
        
    #print("\n")
    
    distance_matrix_platform.append(distance_row)
```

## Compute Release_Date distance_matrices; Release_Date is an ordinal column

```python
release_date = ten_games['Release_Date'].to_list()

release_date
```

### Notice that there are three rank values for column Release_Date.

```python
release_date_ranks = ['older','mid','newer']
```

## Compute the distance matrix of release dates. 

```python
import math

distance_matrix_release_date = []

for i in range(len(release_date)):
    
    distance_row = []
    
    for j in range(len(release_date)):
        
        i_rank_position = -1
        
        j_rank_position = -1
        
        for k in range(len(release_date_ranks)):
            if release_date[i] == release_date_ranks[k]:
                i_rank_position = k
            if release_date[j] == release_date_ranks[k]:
                j_rank_position = k    
        
        distance_ij = (math.fabs(i_rank_position - j_rank_position))/(len(release_date_ranks)-1)
        
        print(i, j, distance_ij)
        
        distance_row.append(round(distance_ij, 2))
    
    print("\n")
    
    distance_matrix_release_date.append(distance_row)
```
