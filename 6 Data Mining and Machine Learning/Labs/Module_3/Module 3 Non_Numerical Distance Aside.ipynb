# Module 3 Non-numerical Distance

```python
# assign 10 colors to l_s.

l_s = ["red", "blue", "green", "green", "blue", "red", "yellow", "purple", "gray", "blue"]
```

# Compute the distance matrix.

```python
import math
```

```python
distance_matrix = []

for i in range(len(l_s)):
    
    distance_row = []
    
    for j in range(len(l_s)):
        #print(str(i) + ", " + str(j))
        if l_s[i] == l_s[j]:
            distance = 0
        else:
            distance = 1
        distance_row.append(round(distance))
        
    #print("\n")
    
    distance_matrix.append(distance_row)
```

```python
# verify the matrix
distance_matrix
```

## Similarity_matrix = 1 - distance_matrix

```python
similarity_matrix = []

for i in range(len(l_s)):
    
    similarity_row = []
    
    for j in range(len(l_s)):
        similarity_row.append(1- distance_matrix[i][j])
    #print("\n")
    
    similarity_matrix.append(similarity_row)
```

```python
# verify the similarity matrix.

print(similarity_matrix)
```
