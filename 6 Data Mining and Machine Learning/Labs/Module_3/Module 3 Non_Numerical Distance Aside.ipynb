# Module 3 Non-numerical Distance

- Compute distances for non-numerical (categorical values)

```python
# assign 10 colors to l_s.

l_s = ["red", "blue", "green", "green", "blue", "red", "yellow", "purple", "gray", "blue"]
```

# Compute the distance matrix. (Categorical)

```python
import math
```

```python
distance_matrix = []

for i in range(len(l_s)):
    
    distance_row = []
    
    for j in range(len(l_s)):
        
        if l_s[i] == l_s[j]:
            distance = 0
        else:
            distance = 1
        distance_row.append(round(distance))
    
    distance_matrix.append(distance_row)
```

### Verify the matrix

```python
distance_matrix
```

## Similarity_matrix = 1 - distance_matrix

```python
similarity_matrix = []

for i in range(len(l_s)):
    
    similarity_row = []
    
    for j in range(len(l_s)):
        similarity_row.append(1- distance_matrix[i][j])
    
    similarity_matrix.append(similarity_row)
```

### Verify the similarity matrix.

```python
print(similarity_matrix)

similarity_matrix
```

# Distance calculation between ordinal values

- assign 10 ranks to l_o
- the order of ranks is indicated in l_o_ranks

```python
l_o = ['B', 'A', 'C', 'D', 'A', 'B', 'B', 'C', 'A','A']

l_o_ranks = ['A', 'B','C', 'D']
```

## An example of distance computation between two ranks.

```python
import math
```

```python
A_position = 0
B_position = 1

distance_AB = (math.fabs(A_position - B_position))/(len(l_o_ranks)-1)

print(distance_AB)
```

```python
A_position = 0
C_position = 2

distance_AC = (math.fabs(A_position - C_position))/(len(l_o_ranks)-1)

print(distance_AC)
```

## An example of distance computation between two end ranks; highest and lowest.

```python
A_position = 0
D_position = 3

distance_AD = (math.fabs(A_position - D_position))/(len(l_o_ranks)-1)

print(distance_AD)
```
