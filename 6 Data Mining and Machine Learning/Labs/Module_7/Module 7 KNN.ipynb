### Define training data and new data.

```python
T = [[6, 9], [4, 6], [8, 7], [4, 4], [6, 2]]		#point 0, point 1, point 2, point 3, point 4
```

```python
labels = [1, 1, 0, 0, 0]	#use 1 for +, 0 for -
```

```python
p = [6, 6] 	#new data point
```

```python
T
```

```python
labels
```

### Define Euclidean distance function.

```python
import math

def distance_euclidean(p, q):
	return math.sqrt((p[0]-q[0]) * (p[0]-q[0]) + (p[1]-q[1])*(p[1]-q[1]))
```

### Compute the distances from new point p to every point in T.

```python
def step_1(T, p):
    
    distances=[]
    
    for i in range(len(T)):
        
        distances.append(distance_euclidean(p, T[i]))
        
    return distances
```

### Rank the distances by sorting.

- used **selection sort** here
- use **any** sorting algorithm to sort the distances and their indices

```python
def step_2(distances):
    
    distance_ranks =[]
    
    for i in range(len(distances)):
        distance_ranks.append(i)

    for i in range(len(distances)-1):
        
        for j in range(i+1, len(distances)):
            
            if distances[i] > distances[j]:
                
                temp_distance = distances[i]
                
                distances[i] = distances[j]
                distances[j] = temp_distance
                
                temp_index = distance_ranks[i]
                
                distance_ranks[i] = distance_ranks[j]
                distance_ranks[j] = temp_index
                
    return distance_ranks
```

Technically, you do not need to sort the whole thing
As long as you return the first k neighbors' indices, that will be good enough.
The sorting here is just for demonstration of using different k values later.

### Find the nearest k neighbors and their labels first. 

- Compute the majority vote next.

```python
def step_3(distances_rank, labels, k):
    
    vote_for_0 = 0
    vote_for_1 = 0
    
    for i in range(k):
        
        if (labels[distances_rank[i]] == 0):
            vote_for_0 = vote_for_0 + 1
        else:
            vote_for_1 = vote_for_1 + 1
            
    if (vote_for_0 > vote_for_1):
        return 0
    else:
        return 1
```

### Demonstrate the case when k = 1.

```python
k = 1

distances = step_1(T, p)
```

```python
distance_ranks = step_2(distances)
```

```python
decision = step_3(distance_ranks, labels, k)
```

```python
print(decision)
```

### Demonstrate the case when k = 3.

```python
k = 3

distances = step_1(T, p)

distance_ranks = step_2(distances)

decision = step_3(distance_ranks, labels, k)

print(decision)
```

### The final prediction/decision.

```python
print(decision)
```

### Demonstrate the case when k = 5.

```python
k = 5

distances = step_1(T, p)

distance_ranks = step_2(distances)

decision = step_3(distance_ranks, labels, k)

print(decision)
```
