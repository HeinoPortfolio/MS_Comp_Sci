# Module 15 Dimensionality

- Demo shows how dimension reduction can be used in visualization.
- uses the Iris dataset from sklearn

### Iris data is used in this demo.

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris_X = iris.data
```

```python
iris_X
```

```python
y = iris.target
```

```python
y
```

### Import PCA.

```python
from sklearn.decomposition import PCA
```

```python
import matplotlib.pyplot as plt
```

### Reduce 4 attributes into 2 PCA components

```python
iris_pca = PCA(n_components = 2).fit_transform(iris_X)
```

```python
iris_pca
```

### Now we can visualizing the data 

```python
plt.scatter(iris_pca[:, 0], iris_pca[:, 1], c = y)
```

### Import TSNE.

```python
from sklearn.manifold import TSNE
```

```python
iris_tsne = TSNE(learning_rate =200).fit_transform(iris_X)
```

```python
iris_tsne
```

### Visualize the data using the two reduced attributes.

```python
plt.scatter(iris_tsne[:, 0], iris_tsne[:, 1], c = y)
```
