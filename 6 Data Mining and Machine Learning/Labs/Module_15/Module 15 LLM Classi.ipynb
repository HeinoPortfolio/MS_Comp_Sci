### Import the needed libraries

- **Note:** May take a few minutes to load this library/ module

```python
from transformers import pipeline
```

### Use the specific model for sentiment classification

- **Note:**      The model has been pre-trained.  No training will be required.
- **Note 2:**    Has a limit of 500 tokens/words. 

```python
classifier = pipeline('sentiment-analysis', model='distilbert-base-uncased-finetuned-sst-2-english')
```

### Use the first 10 reviews.

```python
reviews_10 = df['text'][:10].to_list()
```

```python
reviews_10
```

### Show the number of words in each review.

```python
for v in reviews_10:
    print(len(v.split()))
```

### Tokenize the sentences

```python
from nltk.tokenize import sent_tokenize
```

```python
sentences = sent_tokenize(reviews_10[0])

reviews_10[0]
```

```python
len(sentences)
```

### Use the first 10 sentences

```python
" ".join(sentences[0:10])
```

```python
review_1 = " ".join(sentences[0:10])

classifier([review_1])
```

### Try all 10 reviews


```python
reviews_to_be_used = []

for i in range (len(reviews_10)):

    sentences = sent_tokenize(reviews_10[i])

    if len(sentences) > 10:
        review_short = " ".join(sentences[0:10])
    else:
        review_short = " ".join(sentences)
    
    reviews_to_be_used.append(review_short)
```

```python
reviews_to_be_used
```

### Make some predictions

```python
classifier(reviews_to_be_used)
```
