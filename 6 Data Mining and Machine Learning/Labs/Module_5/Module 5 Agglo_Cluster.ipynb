### Create the followin points for demo.

```python
points = [[52, 28],[25, 95], [36, 64],[20, 94],[52, 27],[20, 96],[39, 65],[34, 90],[75, 15], 
          [74, 21], [90, 43], [60, 12], [67, 20], [30, 92], [61, 31], [8, 74], [51, 32],
          [64, 41], [15, 80], [9, 80]]
```

```python
points
```

### Used for visualization

- method used for visualizing the points

```python
def get_X(P):
    r = []
    for x in P:
        r.append(x[0])
    return r

def get_Y(P):
    r = []
    for x in P:
        r.append(x[1])
    return r
```

### Extract the X and y points

```python
X = get_X(points)

Y = get_Y(points)
```

```python
X
```

```python
Y
```

### Visualize the points

```python
import matplotlib.pyplot as plot

#%matplotlib inline

plot.ylim([0,105])
plot.xlim([0,105])

X = get_X(points)
Y = get_Y(points)

plot.scatter(X, Y)

plot.show()
```

### Import the hierarchical clustering module.

```python
from sklearn.cluster import AgglomerativeClustering
```

### Do the clustering

```python
agg_clustering = AgglomerativeClustering(n_clusters=2, linkage='single').fit(points)
```

### Using **labels_** to see the clustering labels.

- the results
- will show the cluster assignment of each point in the dataset

```python
print(agg_clustering.labels_)
```

### Another example using 4 clusters

```python
agg_clustering_4 = AgglomerativeClustering(n_clusters=4, linkage='single').fit(points)
```

```python
print(agg_clustering_4.labels_)
```

### Plot the cluster

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X, Y, c=agg_clustering.labels_, cmap='rainbow')
plt.title('Agglomerative Clustering')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.show()
```

### Plot the four clusters

```python
plt.scatter(X, Y, c=agg_clustering_4.labels_, cmap='rainbow')
plt.title('Agglomerative Clustering')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.show()
```

### Show the hiarchical tree

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

### Generate the linkage matrix

```python
Z = linkage(points, method='single')
```

### Plot the dendrogram

```python
plt.figure(figsize=(10, 7))

dendrogram(Z)

plt.title('Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')

plt.show()
```
