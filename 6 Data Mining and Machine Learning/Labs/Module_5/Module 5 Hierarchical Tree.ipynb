```python
# import needed modules.
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

### Create the following points for demo.

```python
points = [[8,9],[4,6],[7,5],[5,2],[2,7],[8,4]]
```

### Used for visualization

```python
def get_X(P):
    r = []
    for x in P:
        r.append(x[0])
    return r
```

```python
def get_Y(P):
    r = []
    for x in P:
        r.append(x[1])
    return r
```

```python
X = get_X(points)

Y = get_Y(points)
```

### Visualize the points

```python
import matplotlib.pyplot as plot

#%matplotlib inline

plot.ylim([0,12])
plot.xlim([0,12])

X = get_X(points)
Y = get_Y(points)

plot.scatter(X, Y)

plot.show()
```

### Using labels to denote clusters

```python
labels = ['0', '1', '2', '3', '4', '5']
```

### Visualize clusters

```python
plot.ylim([0,12])
plot.xlim([0,12])
plot.scatter(X, Y, s=100)

for label, x, y in zip(labels, X, Y):
    plot.annotate(label,xy=(x, y), xytext=(5, 5), textcoords='offset points', fontsize=8)
```

### Generate the linkage matrix; single method is using min distance.

-using min distance

```python
Z = linkage(points, method='single')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z)

plt.title('Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')

plt.show()
```

### Generate the linkage matrix; complete method uses max distance.

- uses max distance

```python
Z = linkage(points, method='complete')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z)

plt.title('Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

```python
# Generate the linkage matrix; using the average distance.
Z = linkage(points, method='average')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
