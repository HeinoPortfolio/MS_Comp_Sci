### Create the following points for demo.

```python
points = [[8,9],[4,6],[7,5],[5,2],[2,7],[8,4]]
```

```python
### Define a hamming distance function (Manhattan distance)
```

```python
import math

def my_distance(p, q):
    return math.fabs(p[0] - q[0]) + math.fabs(p[1] - q[1])
```

### Define a distance function between two clusters using min.

```python
def my_distance_min(C1, C2):
    
    distance = 1000
    
    for p in C1:
        for q in C2:
            if my_distance(p, q) < distance:
                distance = my_distance(p, q)
    return distance
```

### Cluster A has only one point, and Cluster B has only one point as well. 
### Verify my_distance_min().

```python
A = [[8,9]]
B = [[4,6]]

print(my_distance_min(A, B))
```

### Merge two clusters using the min distance between any two clusters.

```python
cluster_i = -1
cluster_j = -1

shortest_distance = 1000

for i in range(len(points) - 1):
    for j in range(i+1, len(points)):
        A = [points[i]]
        B = [points[j]]
        min_distance = my_distance_min(A, B)
        if min_distance < shortest_distance:
            cluster_i = i
            cluster_j = j
            shortest_distance = min_distance

print("merge " + str(cluster_i) + " and " + str(cluster_j))
```

### The distance between two clusters is defined as the max distance.

```python
def my_distance_max(C1, C2):
    
    distance = 0
    
    for p in C1:
        for q in C2:
            if my_distance(p, q) > distance:
                distance = my_distance(p, q)
    return distance
```

### Merge two clusters using the max distance between any two clusters.

```python
cluster_i = -1
cluster_j = -1

shortest_distance = 1000

for i in range(len(points) - 1):
    for j in range(i+1, len(points)):
        A = [points[i]]
        B = [points[j]]
        min_distance = my_distance_max(A, B)
        if min_distance < shortest_distance:
            cluster_i = i
            cluster_j = j
            shortest_distance = min_distance

print("merge " + str(cluster_i) + " and " + str(cluster_j))
```

Distance between C1 cluster and C2 cluster using **min** distance.

```python
C1 = [[8,9],[4,6]]
C2 = [[7,5],[5,2]]

print(my_distance_min(C1, C2))
```

Distance between C1 cluster and C2 cluster using **max** distance.

```python
# Distance between C1 cluster and C2 cluster using max distance.
C1 = [[8,9],[4,6]]
C2 = [[7,5],[5,2]]
print(my_distance_max(C1, C2))
```

### The distance between two clusters is defined as the average distance between any two points from different clusters

```python
def my_distance_avg(C1, C2):
    
    distance = 0
    
    for p in C1:
        for q in C2:
            distance = distance + my_distance(p, q)
    return distance/(len(C1)*len(C2))
```

### Distance between C1 cluster and C2 cluster using avg(mean) distance.

```python
C1 = [[8,9],[4,6]]
C2 = [[7,5],[5,2]] 

print(my_distance_avg(C1, C2))
```

### Hamming distance between any two points from C1 and C2. 

```python
for p in C1:
    for q in C2:
        print(my_distance(p, q))
```
