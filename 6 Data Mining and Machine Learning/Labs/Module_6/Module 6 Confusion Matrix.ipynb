### Show how to compute confusion matrix

- confusion matrix

The 20 true labels. A can be seen as positive. B can be seen as negative.

```python
true_labels = ['A', 'A', 'A', 'A', 'A', 'A', 'A', 'A','A', 'A', 'B', 'B', 'B','B', 'B', 'B', 'B', 'B', 'B', 'B']
print(true_labels)
```

The 20 predicted labels.

```python
predicted_labels = ['A', 'B', 'A', 'B', 'B', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'A', 'A', 'A', 'B', 'B', 'B', 'B']

print(predicted_labels)
```

Some initializations

```python
num_of_true_A = 0
num_of_true_B = 0

num_of_predicted_A = 0
num_of_predicted_B = 0
```

```python
for x in true_labels:
    if x == 'A':
        num_of_true_A = num_of_true_A + 1
    elif x == 'B':
        num_of_true_B = num_of_true_B + 1
    else:
        print("Error!")
        break

print(num_of_true_A)
print(num_of_true_B)
```

```python
for x in predicted_labels:
    if x == 'A':
        num_of_predicted_A = num_of_predicted_A + 1
    elif x == 'B':
        num_of_predicted_B = num_of_predicted_B + 1
    else:
        print("Error!")
        break
        
print(num_of_predicted_A)
print(num_of_predicted_B)
```

Show true_positive, true negative, false positive, and false negative.

```python
true_positive = 0
true_negative = 0
false_positive = 0
false_negative = 0

for i in range(len(true_labels)):
    if (true_labels[i] == predicted_labels[i]) and (true_labels[i] == 'A'): 
        true_positive = true_positive + 1
    elif (true_labels[i] == predicted_labels[i]) and (true_labels[i] == 'B'): 
        true_negative = true_negative + 1
    elif (true_labels[i] != predicted_labels[i]) and (true_labels[i] == 'A'): 
        false_negative = false_negative + 1
    elif (true_labels[i] != predicted_labels[i]) and (true_labels[i] == 'B'): 
        false_positive = false_positive + 1    
    else:
        print("Error!")
        break

print(true_positive)
print(true_negative)
print(false_positive)
print(false_negative)
```

```python
print(f"true_positive: {true_positive}")
```

### Compute accuracy

```python
print(f"accuracy: {(true_positive+true_negative)/(true_positive+true_negative+false_positive+false_negative)}")
```

### Compute precision

```python
precision = (true_positive)/(true_positive+false_positive)

print(precision)
```

### Compute recall

```python
print(f"recall: {(true_positive)/(true_positive+false_negative)}")

recall = (true_positive)/(true_positive+false_negative)

print(recall)
```

```python
### Compute F1 score
```

```python
f1 = 2 * ((precision*recall)/(precision+recall))

print(f1)
```
