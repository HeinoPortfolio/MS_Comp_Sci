### Identify the distinct items contained in baskets

```python
def generate_unique_items(T):
    
    C = []
    
    for t in T:
        for x in t:
            if x not in C:
                C.append(x)
    
    C.sort()
    
    return C
```

### Generate the itemset

```python
def generate_C1(U ,T):
   
    d = {}
    
    for x in U:
        count = 0
        for t in T:
            if x in t:
                count = count + 1
        d[x] = count
    
    return d
```

### Filter out the items that satisfy the min support.

```python
def generate_L1(C_1, min_support):
    
    d = C_1
   
    L1={}
   
    for k in d.keys():
        #print(k)
        if d[k] >=  min_support:
            L1[k] = d[k]        
    return L1
```

### Generate item sets with two items.

```python
def generate_C2_pairs(L1):
    
    pairs = []
    
    key_list = list(L1.keys())
    
    for i in range(len(L1.keys())-1):
        
        for j in range(i+1, len(key_list)):
            
            pair = [key_list[i], key_list[j]]
            pair.sort()
            pairs.append(pair)
            
    return pairs
```

### Utility function

```python
def included(t, pair):
    
    subset = True
    
    for x in pair:
        if x not in t:
            subset = False
    return subset
```

### Generate item sets with two items and their support/frequency

```python
def generate_C2(pairs ,T):
    
    d = {}
    
    for pair in pairs:
        count = 0
        for t in T:     
            if included(t, pair):
                count = count + 1
        d[tuple(pair)] = count
        
    return d
```
